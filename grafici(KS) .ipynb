```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import matplotlib.patches as mpatches
from scipy.optimize import curve_fit
```

```python
Z, N, E, err_e = np.loadtxt("./KS/Energia_variazionali_finali.dat", usecols=(0,1,2,3), delimiter=' ', unpack='true')

for i in range (0, len(Z)):
    print(Z[i], "  " ,N[i], "  " , E[i], "  " , err_e[i])
```

```python
fig, (f1) = plt.subplots(1,1,figsize=(10,6))

x, f = np.loadtxt("./KS/risultati(Z=N=1)/autofunzioni/n_1l_0ms_0.dat", usecols=(0,1), delimiter='	', unpack='true')
f1.plot(x,f, '-r', label='Z=N=1 (1s)')


f1.plot(x,2*np.exp(-x)*x,'--b', label='H (1s)')

f1.set_xlabel("$r$",fontsize=15)
f1.set_ylabel("$r*R(r)$",fontsize=15)

plt.locator_params(axis='y', nbins=5)
plt.locator_params(axis='x', nbins=6)

f1.tick_params(axis='both', which='major', labelsize=14.5)


leg = f1.legend(fontsize=14);
f1.set_xlim(0,8)
f1.set_ylim(-0.1,1)
f1.grid(True)
plt.show()
```
