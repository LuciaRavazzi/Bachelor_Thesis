```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import matplotlib.patches as mpatches
from scipy.optimize import curve_fit
```

```python
fig, (f1) = plt.subplots(1,1,figsize=(12,7))

x, f = np.loadtxt("./Thomas-Fermi/TF/Density/Density(Z=1,N=1).dat", usecols=(0,1), delimiter=' ', unpack='true')
f1.plot(x,np.log(f), '-r', label='Z=N=1')

x, f = np.loadtxt("./Thomas-Fermi/TF/Density/Density(Z=2,N=2).dat", usecols=(0,1), delimiter=' ', unpack='true')
f1.plot(x,np.log(f), '--b', label='Z=N=2')

x, f = np.loadtxt("./Thomas-Fermi/TF/Density/Density(Z=18,N=18).dat", usecols=(0,1), delimiter=' ', unpack='true')
f1.plot(x,np.log(f), '-.g', label='Z=N=10')


f1.set_xlabel("$r$",fontsize=15)
f1.set_ylabel("$n(r)$",fontsize=15)
f1.tick_params(axis='both', which='major', labelsize=14.5)

leg = f1.legend(fontsize=14);

plt.locator_params(axis='y', nbins=5)
plt.locator_params(axis='x', nbins=6)

plt.savefig('sample.pdf')
plt.grid(True)
plt.show()
```

```python
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes
from mpl_toolkits.axes_grid1.inset_locator import mark_inset

fig, (f2) = plt.subplots(1,1,figsize=(10,10))
Z, mu, k, p, e = np.loadtxt("./Thomas-Fermi/TF(pt.2)/Dati.dat", usecols=(0,1,2,3,4), delimiter=' ', unpack='true')

f2.plot(Z,k+p+e, color="red",  label="Energia totale TF")

f2.plot(Z,p,  '--g',  label='Energia potenziale')

f2.plot(Z,k, 'c' , marker=".",label='Energia cinetica')

f2.plot(Z,e, '-.',  label='Energia interazione e/e')

leg = f2.legend(fontsize=14)
f2.set_xlabel("$N=Z$",fontsize=15)
f2.tick_params(axis='both', which='major', labelsize=14.5)
f2.grid(True)

plt.locator_params(axis='y', nbins=4)
plt.locator_params(axis='x', nbins=4)


axins = zoomed_inset_axes(f2, 42, loc=3)
axins.plot(Z,k+p+e, color="red",  label="Energia TF")
axins.plot(Z,p,  '--g',  label='Energia potenziale')
axins.plot(Z,k, 'c' , marker=".",label='Energia cinetica')
axins.plot(Z,e, '-.',  label='Energia interazione e/e')

axins.locator_params(axis='y', nbins=2)
axins.locator_params(axis='x', nbins=4)

axins.set_xlim(0.9, 3) # apply the x-limits
axins.set_ylim(-300, 400) # apply the y-limits
plt.yticks(visible=False)
plt.xticks(visible=False)
axins.grid(True)
mark_inset(f2, axins, loc1=1, loc2=2, fc="none", ec="0.9")

plt.show()
```
